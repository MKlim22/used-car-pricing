```python
import pandas as pd
import numpy as np
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

df = pd.read_csv(url, header=None)
```

```python
df.to_csv("data/auto.csv", index=False, header=False)

print("Saved to data/auto.csv")
```

```python
df.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
```

```python
df.describe(include="all")
```

```python
df.replace("?", np.nan, inplace=True)
df.isnull().sum()
```

```python
numeric_cols = ["normalized-losses","bore","stroke","horsepower","peak-rpm","price"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df = df.dropna(subset=["price"])

df['normalized-losses']=df['normalized-losses'].fillna(df['normalized-losses'].median())

df['num-of-doors']=df['num-of-doors'].fillna('four')

for col in ['bore', 'stroke', 'horsepower', 'peak-rpm']:
    df[col]=df[col].fillna(df[col].mean())
```

```python
numeric_df = df.select_dtypes(include="number")
corr_matrix = numeric_df.corr()
corr_with_price = corr_matrix['price'].sort_values(ascending=False)

corr_with_price
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Price vs drive-wheels
plt.figure(figsize=(6,4))
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.title('Price vs Drive Wheels')
plt.show()
```

### Price vs Drive Wheels

Cars with rear-wheel drive tend to have higher median price than front-wheel drive vehicles. This suggests that drivetrain configuration has a noticeable impact on pricing.

```python
# Prise vs body-style
plt.figure(figsize=(8,4))
sns.boxplot(x='body-style', y='price', data=df)
plt.title('Price vs Body Style')
plt.show()
```

### Price vs Body style

Convertibles and hardtops generally show higher prices compared to other body styles, while hatchback and wagons are typically more affordable.

```python
# Engine size vs price
plt.figure(figsize=(6,4))
sns.regplot(x='engine-size', y='price', data=df)
plt.title('Engine Size vs Price')
plt.show()
```

### Engine Size vs Price

A strong positive linear relationship is visible between engine size and price, indicating engine size as a key predictor of vehicle value.

```python
# Horse power vs price
plt.figure(figsize=(6,4))
sns.regplot(x='horsepower', y='price', data=df)
plt.title('Horsepower vs Price')
plt.show()
```

### Horsepower vs Price

A clear positive relationship is observed between horsepower and price, with higher-performance vehicles generally commanding higher prices.
